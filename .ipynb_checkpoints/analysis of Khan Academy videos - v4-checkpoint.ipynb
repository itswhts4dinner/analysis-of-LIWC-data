```python
import pandas as pd
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

# https://pingouin-stats.org
import pingouin as pg
```

```python
## Import all instruction video csv files
instructional_data = sorted(glob('./transcripts/instructional/*.csv',recursive=True))
instructional_data
```

```python
#Import all edutainment video csv files
edutainment_data = sorted(glob('./transcripts/edutainment/*.csv',recursive=True))
edutainment_data
```

```python
## Make a dataframe for instructional and edutainment videos

LIWC_instructional = pd.concat((pd.read_csv(file).assign(channel_name=file)
          for file in instructional_data))
LIWC_edutainment = pd.concat((pd.read_csv(file).assign(channel_name=file)
          for file in edutainment_data))
```

```python
columns_of_interest = {'Analytic', 'Tone', 'Clout', 'Authentic'}
```

```python
## Test instructional data for normality
for i in columns_of_interest:
    print("Results for", i)
    x_instructional = LIWC_instructional[i]
    print(pg.normality(x_instructional))
    pg.qqplot(x_instructional, dist='norm')
    #scipy.stats.probplot(x_instructional, dist="norm", plot=plt)
    plt.show()
```

```python
## Test edutainment data for normality
for i in columns_of_interest:
    print("Results for", i)
    x_edutainment = LIWC_edutainment[i]
    pg.qqplot(x_edutainment, dist='norm')
    print(pg.normality(x_edutainment))
    plt.show()
```

```python
## Make the instructional group histograms
f, ax = plt.subplots(figsize=(8,8))
plt.title('Instructional Video Analytic Scores')
sns.histplot(LIWC_instructional.Analytic, bins=20,  color="indigo")

#####
f, ax = plt.subplots(figsize=(8,8))
plt.title('Instructional Video Clout Scores')
sns.histplot(LIWC_instructional.Clout, color="indigo")

#####
f, ax = plt.subplots(figsize=(8,8))
plt.title('Instructional Video Authentic Scores')
sns.histplot(LIWC_instructional.Authentic, color="indigo")

#####
f, ax = plt.subplots(figsize=(8,8))
plt.title('Instructional Video Tone Scores')
sns.histplot(LIWC_instructional.Tone, color="indigo")
```

```python
#Make the edutainment group histograms
f, ax = plt.subplots(figsize=(8,8))
plt.title('Edutainment Video Analytic Scores')
sns.histplot(LIWC_edutainment.Analytic, bins=20, color="green")

#####
f, ax = plt.subplots(figsize=(8,8))
plt.title('Edutainment Video Clout Scores')
sns.histplot(LIWC_edutainment.Clout, color="green")

#####
f, ax = plt.subplots(figsize=(8,8))
plt.title('Edutainment Video Authentic Scores')
sns.histplot(LIWC_edutainment.Authentic, color="green")

#####
f, ax = plt.subplots(figsize=(8,8))
plt.title('Edutainment Video Tone Scores')
sns.histplot(LIWC_edutainment.Tone, color="green")
```
