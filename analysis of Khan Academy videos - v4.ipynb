```python
import pandas as pd
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# https://pingouin-stats.org
import pingouin as pg
```

```python
## Import all instruction video csv files
instructional_data = sorted(glob('./transcripts/instructional/*.csv',recursive=True))
instructional_data
```

```python
#Import all edutainment video csv files
edutainment_data = sorted(glob('./transcripts/edutainment/*.csv',recursive=True))
edutainment_data
```

```python
## Make a dataframe for instructional and edutainment videos

LIWC_instructional = pd.concat((pd.read_csv(file).assign(channel_name=file)
          for file in instructional_data))
LIWC_edutainment = pd.concat((pd.read_csv(file).assign(channel_name=file)
          for file in edutainment_data))

```

```python
x_instructional = LIWC_instructional['Analytic']

pg.qqplot(x_instructional, dist='norm')

f, ax = plt.subplots(figsize=(8,8))
plt.title('Instructional Video Analytic Scores')
sns.histplot(LIWC_instructional.Analytic, bins=20,  color="indigo")


pg.normality(x_instructional)
```

```python
x_edutainment = LIWC_edutainment['Analytic']

##Make a QQ plot
pg.qqplot(x_edutainment, dist='norm')


#Make the histogram
f, ax = plt.subplots(figsize=(8,8))
plt.title('Edutainment Video Analytic Scores')
sns.histplot(LIWC_edutainment.Analytic, bins=20, color="green")

pg.normality(x_edutainment)
```
